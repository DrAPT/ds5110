2) **Work with Athena**  
This is a tool for writing/running/storing SQL queries.

Follow the tutorial to build a table for log data, called *elb_logs*.  It is very cool and fun, and at the end, you will have created a table from data in an S3 bucket.  They also show the code generated by the wizard.  

```python
# count the number of records in the table  
SELECT count(*)  
FROM elb_logs;  

# fetch the first 10 records from the table  
SELECT *   
FROM elb_logs  
LIMIT 10;  
```

By the way, if you just want to look at some records in the table, it is ALWAYS a good idea add a LIMIT statement.  For SELECT queries against a massive table, it might take forever to finish.

3) **Work with Glue**  
This is an amazing ETL tool.  


S3 buckets can hold all sorts of data at various levels of structure. For unstructured data such as web pages, SQL won't work directly.  Glue allows users to set up crawlers that crawl the data and discover schema.  For example, running a crawler on a semistructured format such as JSON will produce a structured table (fields and records).

After Glue produces the table, Athena can be used to query it.  Glue also generates the code used to create the table, which for large datasets, can be formidible.
